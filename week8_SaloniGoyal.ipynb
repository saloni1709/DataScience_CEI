# 🧠 Single Agent Pipeline Project

## Problem Statement
Build a **Single-Agent Smart Assistant** that:
- Understands user queries
- Routes tasks based on intent
- Uses tools when required
- Returns structured JSON output

### The agent should handle:
- Math queries → Calculator Tool
- Keyword extraction → Keyword Tool
- General queries → Direct response

---
### 🛠️ What You Need to Implement
- Agent logic
- Conditional routing
- Tool integration
- Basic error handling

### 🚀 Bonus
- Improve routing
- Add logging
- Add more tools


````python
# 🛠️ TOOL 1: Calculator

def calculator(expression: str) -> str:
    """Evaluate a mathematical expression."""
    try:
        return str(eval(expression))
    except Exception:
        return "Error in calculation"
````

````python
# 🛠️ TOOL 2: Keyword Extractor

def extract_keywords(text: str) -> list:
    """Extract keywords from text."""
    try:
        words = text.split()
        keywords = list(set([w.lower() for w in words if len(w) > 4]))
        return keywords[:5]
    except Exception:
        return []
````

## 🤖 Implement Agent Logic Below

👉 Use conditional routing:
- If query contains "calculate" → use calculator
- If query contains "keywords" → use keyword extractor
- Else → general response

````python
# 🤖 AGENT FUNCTION

def agent(user_query: str):
    """Decides which function should handle the user's request."""

    try:
        query = user_query.lower()

        # Check if the user wants to perform a calculation
        if "calculate" in query:
            expression = user_query.replace("calculate", "").strip()
            answer = calculator(expression)

            if answer == "Error in calculation":
                return {
                    "type": "error",
                    "result": answer
                }

            return {
                "type": "calculation",
                "result": answer
            }

        # Check if the user wants keyword extraction
        if "keywords" in query:
            text = (
                user_query.replace("Extract keywords from", "")
                .replace("extract keywords from", "")
                .strip()
            )

            keywords = extract_keywords(text)

            return {
                "type": "keywords",
                "result": keywords
            }

        # Default response for other queries
        return {
            "type": "general",
            "result": "No matching tool found for this request."
        }

    except Exception as error:
        return {
            "type": "error",
            "result": str(error)
        }
````

## 📦 Expected Output Format

```
{
  "type": "calculation / keywords / general / error",
  "result": ...
}
```

````python
# 🧪 Test Cases

queries = [
    "Calculate 20 + 5",
    "Extract keywords from Artificial Intelligence is transforming industries",
    "What is machine learning?"
]

for q in queries:
    print("Query:", q)
    print("Response:", agent(q))
    print("-" * 50)
````

````python
# 🎯 Interactive Mode

while True:
    user_input = input("Enter query (type 'exit' to stop): ")
    if user_input.lower() == "exit":
        break
    print("Response:", agent(user_input))
````
